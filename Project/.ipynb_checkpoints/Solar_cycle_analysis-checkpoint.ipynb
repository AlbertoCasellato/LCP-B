```r
# Solar Cycles Data Analysis
# Solar Cycle Inference Script via MCMC (timedeppar lib)
# -----------------------------------------------
# Authors:  Francesco Pazzocco - ALberto Casellato - Elias Bonasera - Nicola Gabrin
# Date:     2025‑05‑08
# ------------------ Libraries ------------------
library(readxl)
library(timedeppar)
library(dplyr)
library(tidyr)
library(ggplot2)
```

```r
# ------------------ Data Import & Preparation ------------------

# ------------------------   Harmonics   ------------------------
# The harmonics parameter grid must follow the structure:
#       Period(years) | sigma_period | Amplitude | sigma_amplitude | Phase | sigma_phase | Significance | Inferred

# Notes:
# - "Significance" indicates how well a harmonic fits the target data.
# -  Harmonics with Significance >= significance_threshold are treated as "fixed" (inferred)
# -  Frequencies are returned as a list of lists with:
#       list(fixed = TRUE/FALSE, omega = 2 * pi / Period)
# -  Inferred --------> [0 or 1]: 0= non fixed   1= fixed  this will be the method to select if an harmonics is fixed or has to be inferred

load_harmonics   <- function(path = "./data/", filename = "harmonics_grid.xlsx", significance_threshold = 0.9, skip = 4, col_names = TRUE) {
    harmonics_df <- read_excel(paste0("./data/", filename), skip = skip, col_names = col_names)      # harmonics value on raw(6) raw(5) is metadata
    return(harmonics_df)
}


# ------------------   Dataset Age_BP - d18O   ------------------ 

# load_solar_data()       # to load the dataset use in workflow section
# trim_solar_data()       # to cut the dataset intervals (eg. 0-30k)

load_solar_data <- function(path = "./data/", filename = "solar_data.xlsx", skip = 4, col_names = TRUE) {
  df_solar <- read_excel(paste0(path, filename), skip = skip, col_names = col_names)
  df_solar <- df_solar %>%
              mutate(
                Year = 1950 - Age,                                 # Conversion: BP <-> standard year
                Cycle = scale(d18O, center = TRUE, scale = FALSE)  # Center only ! Verified it is consistent with the harmonics amplitudes !
              ) %>%
              arrange(desc(Age)) %>% 
              select(Year, Cycle)
  return(df_solar)
}

trim_solar_data <- function(df_solar, lower = 0, upper = 30000) {
  cutoff_year_min  <- 1950 - upper
  cutoff_year_max  <- 1950 - lower  
  df_solar_trimmed <- df_solar %>%
    filter(Year >= cutoff_year_min & Year <= cutoff_year_max)  

  return(df_solar_trimmed)
}

# ------------------   Model Tools   ------------------

# ------------------  Loglikelihood  ------------------
loglikeli <- function(param, data, harmonics) {
  xi <- param$xi[, 2]
  t  <- cumsum(c(data$Year[1], xi))    # t  <- c(data$Year[1], cumsum(xi) + data$Year[1])
  y  <- 0
  for (i in 1:nrow(harmonics)) {
                    A   <- param[[paste0("A", i)]]                                                       # Amplitude
                    phi <- param[[paste0("phi", i)]]                                                     # Phase
                    if (harmonics[i, ]$inferred == 1) { omega <- 2 * pi / harmonics[i, ]$Period}         # Fixed frequency
                    else { omega <- param[[paste0("omega", i)]] }                                        # Inferred frequency
                    y <- y + A * cos(omega * t + phi)
                  }  
  loglik <- sum(dnorm(data$Cycle, mean = y, sd = exp(param$sigma_y), log = TRUE))
  return(loglik)
}

# ------------------  Inizialization  ------------------
initialize_params <- function(df, harmonics) {
  observed_intervals <- diff(df$Year)
  xi_time            <- head(df$Year, -1)     # Previus version:  xi_time <- df$Year[1:(nrow(df) - 1)] 
  xi_initial_matrix  <- cbind(                # cbind(xi_time, randOU(...)$y) == MATRIX (n−1)×2 format: [time; epsilon]
    xi_time,                                  # vector of (n-1) year
    randOU(                                   # Structure  randOU(..., t = xi_time)$y
            mean  = mean(observed_intervals), 
            sd    = xi_sd,                    # MUST BE length(xi_time) == length(randOU(...)$y)
            gamma = xi_gamma,
            t     = xi_time)$y
          )
  # Same for all harmonics
  param.ini <- list(xi = xi_initial_matrix, sigma_y  = sigma_y ) # check sigma_y !!!!!
                  
  # Initialize harmonics from grid
    for (i in 1:nrow(harmonics)) {
    param_A   <- paste0("A", i)
    param_phi <- paste0("phi", i)
    row       <- harmonics[i, ]
    param.ini[[param_A]]   <- row$Amplitude
    param.ini[[param_phi]] <- row$Phase
    if (harmonics[i, ]$inferred == 0) { param.ini[[paste0("omega", i)]] <- 2 * pi / row$Period }     # Non-fixed harmonics
  }
  # OU parameters
  param.ou.ini <- c(
                      xi_mean  = mean(observed_intervals),
                      xi_sd    = xi_sd,
                      xi_gamma = xi_gamma
                  )
  return(list(param.ini = param.ini, param.ou.ini = param.ou.ini))
}

# ------------------  Inference  ------------------
run_inference <- function(df, param.ini, param.ou.ini, harmonics,
                          n.iter     = 50000,
                          n.adapt    = 10000,
                          n.int      = 10,
                          stepsize   = 0.1,
                          oustepsize = 0.1,
                          verbose    = 0) {
  control <- list(
                   n.iter           = n.iter,     # Number of iterations
                   n.adapt          = n.adapt,    # Burn-in
                   n.interval       = n.int,      # Sub-intervals of time domain
                   scale.prop.const = stepsize,   # how much it move in each step
                   scale.prop.ou    = oustepsize  # same but for OU parameters
                 )
# Parameters ranges all ranges are builded as [x-3sigma_x; x+ 3sigma_x] where x is the generic parameter with its error sigma
  # Precompute statistics
  observed_intervals <- diff(df$Year)
  mean_xi            <- mean(observed_intervals)
  sd_xi              <- sd(observed_intervals)
  sigma_y_init       <- sd(df$Cycle, na.rm = TRUE)
  lower_sy           <- max(1e-3, sigma_y_init - 3 * sigma_y_init)
  upper_sy           <- sigma_y_init + 3 * sigma_y_init
  # Initialization param.range + param.log    
  param.range <- list(sigma_y = c(lower_sy, upper_sy), xi = c(max(1e-2, mean_xi - 3 * sd_xi), mean_xi + 3 * sd_xi)) # Lower bound to avoid errors
  param.log   <- c(sigma_y = TRUE)  # Noise is on log scale
  # Harmonics param.range both fixed and inferred
  for (i in 1:nrow(harmonics)) {
    row         <- harmonics[i, ]
    param_A     <- paste0("A", i)
    param_phi   <- paste0("phi", i)
    param_omega <- paste0("omega", i)
      
    if (harmonics[i, ]$inferred == 0) {                      # Ranges for inferred harmonics
                param_omega <- paste0("omega", i)     # Error propagation = sigma_period * ( 2 * pi / Period ^ 2 )
                T           <- row$Period
                sigma_omega <- abs(2 * pi / T^2) * row$sigma_period                       
                omega_mean  <- 2 * pi / T
    param.range[[param_omega]]  <- c(omega_mean - 3 * sigma_omega, omega_mean + 3 * sigma_omega )
    param.log[param_omega]      <- TRUE
    }
  param.range[[param_A]]        <- c(max( 0.01, row$Amplitude - 3 * row$sigma_amplitude), row$Amplitude + 3 * row$sigma_amplitude )
  param.range[[param_phi]]      <- c(row$Phase - 3 * row$sigma_phase, row$Phase + 3 * row$sigma_phase)  
  }  
  # Fix OU parameters for intervals (xi)
  #  !!!!! VERIFY IF IT IS NEEDED OR NOT  !!!!!
#  param.ou.fixed <- c(
#    xi_mean = param.ou.ini$xi_mean,
#    xi_gamma = param.ou.ini$xi_gamma
#  )
        
        
  res <- infer.timedeppar(
            loglikeli      = loglikeli,
            data           = df,
            param.ini      = param.ini,
            param.range    = param.range,
            param.log      = param.log,
            param.ou.ini   = param.ou.ini,
#            param.ou.fixed = param.ou.fixed,    # just decomment / comment in case needed / not needed
            n.iter         = control$n.iter,
            control        = control,
            verbose        = verbose
          )
  return(res)
}

# ------------------   Visualization   ------------------
plot_all <- function(res, data, title = "Results") {
  # 1) timedeppar diagnostics
  plot(res, type = c("traces", "marginals", "time-series", "pairs"), main = paste(title, "- diagnostics"))

  # inferred vs observed times
  xi_mat     <- res$sample.param.timedep$xi
  xi_samples <- as.matrix(xi_mat[-1, , drop = FALSE])
  t_matrix   <- t(apply(xi_samples, 1, function(xi_vals) {
    c(data$Year[1], data$Year[1] + cumsum(xi_vals))
  }))
  q          <- apply(t_matrix, 2, quantile, probs = c(0.025, 0.5, 0.975))
  df_plot    <- tibble::tibble(
    Year   = data$Year,
    q025   = q[1, ],    # Lower bound of 95% credible interval
    median = q[2, ],    # Inferred median time
    q975   = q[3, ]     # Upper bound of 95% credible interval
  )
  
  ggplot(df_plot, aes(Year)) +
                               geom_ribbon(aes(ymin = q025, ymax = q975, fill = "95% Credible Interval"), alpha = 0.5) +
                               geom_line(aes(y = median, color = "Inferred Median"), linewidth = 0.5) +
                               geom_line(aes(y = Year, color = "Observed Year"), linetype = 2) +
    scale_fill_manual(name = "Uncertainty",values = c("95% Credible Interval" = "grey80")) +
    scale_color_manual(name = "Series", values = c("Inferred Median" = "black", "Observed Year"   = "red")) +
    labs(title = title, y = "Time") +
    theme_minimal() +
    theme(
          legend.position.inside = c(0.1, 0.9),
          legend.justification   = c(0, 1),
          legend.background      = element_rect(fill = "white", color = NA, linewidth = 0),
          legend.key             = element_rect(fill = NA)
         )
}
```

```r
# -------------   Pipeline   -------------
# -----------  Load Harmonics  -----------
harmonics      <- load_harmonics(filename = "solar_harmonics.xlsx", significance_threshold = 0.9)

# -----------  Load & Trim Data -----------
df_solar_full  <- load_solar_data()
df_solar_0_30  <- trim_solar_data(df_solar_full, lower = 0, upper = 30000)  

# -----------   Initialization  -----------
init           <- initialize_params(df_solar_0_30)
param.ini      <- init$param.ini
param.ou.ini   <- init$param.ou.ini

# -----------  Run Inference  -----------
result         <- run_inference(df_solar_0_30, param.ini, param.ou.ini, n.iter = 50000, n.adapt = 10000, n.int = 25)
```

```r
# -----------  Plot Results  -----------
plot_all(result, df_solar_0_30, title = "Harmonic Inference: 0–30000")
```
